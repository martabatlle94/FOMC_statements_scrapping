```python
from selenium import webdriver 
from selenium.webdriver.common.by import By 
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support import expected_conditions as EC 
from selenium.common.exceptions import TimeoutException
from datetime import datetime
import pandas as pd
pd.set_option('display.max_rows', None)
browser = webdriver.Chrome()
```

```python
dates= pd.read_excel("dates_excel.xlsx")
```

```python
dates['date']= pd.to_datetime(dates['date'], format='%Y-%M-%D')
```

```python
dates['Year']= dates['date'].dt.strftime("%Y")
dates['Month']= dates['date'].dt.strftime("%m")
dates['Day']= dates['date'].dt.strftime("%d")


dates['date']= dates['Year']+dates['Month']+dates['Day']
```

```python
dates1= list(dates.date[0:132])         #they changed the format over time so we need 3 different scrappings
dates2= list(dates.date[132:162])
dates3= list(dates.date[162:])
```

```python
year2= list(dates.Year[132:162])
year3= list(dates.Year[162:])
```

```python
browser.get("https://www.federalreserve.gov/monetarypolicy/materials/")
```

```python
speeches=[]
alldates=[]

for i in dates1:
    browser.get(f"https://www.federalreserve.gov/newsevents/pressreleases/monetary{i}a.htm")
    statement = browser.find_elements_by_id("content")
    statement_text = [x.text for x in statement]
    speeches.append(statement_text)
    alldates.append(i)
```

```python
for i, j in zip(dates2, year2):
    browser.get(f"https://www.federalreserve.gov/boarddocs/press/monetary/{j}/{i}/default.htm")
    statement = browser.find_elements_by_css_selector('body')
    statement_text = [x.text for x in statement]
    speeches.append(statement_text)
    alldates.append(i)
```

```python
for i, j in zip(dates3, year3):  
    browser.get(f"https://www.federalreserve.gov/boarddocs/press/general/{j}/{i}/default.htm")
    statement = browser.find_elements_by_css_selector('body')
    statement_text = [x.text for x in statement]
    speeches.append(statement_text)
    alldates.append(i)
```

```python
speechesdf=pd.DataFrame(speeches).reset_index(drop=True)

datedf=pd.DataFrame(alldates).reset_index(drop=True)

final=speechesdf.merge(datedf, left_index=True, right_index=True)
```
